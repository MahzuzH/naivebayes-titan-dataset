```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
titan = pd.read_csv("train.csv")
```

```python
titanA = titan.drop(columns=['Ticket','Name','Cabin','PassengerId'])
titanB = titan[['Ticket','Name','Cabin']]
print(titanA)
```

```python
Encoder = LabelEncoder()
titanA['Sex'] = titanA['Sex'].astype('category')
titanA['Embarked'] = titanA['Embarked'].astype('category')

for col in ['Sex', 'Embarked']:
    titanA[col] = Encoder.fit_transform(titanA[col])
titanA.head(6)
```

```python
null_counts = titanA['Age'].isnull().sum()
total_nulls = null_counts.sum()
print("Total null values:", total_nulls)
```

```python
mean_age = titanA['Age'].mean()
titanA['Age'].fillna(mean_age, inplace=True)
titanA
```

```python
null_counts = titanA['Age'].isnull().sum()
total_nulls = null_counts.sum()
print("Total null values:", total_nulls)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score
```

```python
X = titanA.drop('Survived', axis=1)
y = titanA['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Build a Gaussian Classifier
model = GaussianNB()

# Model training
model.fit(X_train, y_train)

# Predict Output
predicted = model.predict(X_test)

print("Actual Value   :", y_test)
print("Predicted Value:", predicted)
```

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    recall_score,
    precision_score
)

y_pred = model.predict(X_test)
accuray = accuracy_score(y_pred, y_test)
prediction = precision_score(y_pred, y_test)
recall = recall_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("Prediction:", prediction)
print("Recall:", recall)
print("F1 Score:", f1)
```
